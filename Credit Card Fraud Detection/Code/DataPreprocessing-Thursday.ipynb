```python
# Creata a distance variable for place of transaction from house address of card holder
def haversine_vectorized(lat1, lon1, lat2, lon2):
    """
    Compute the Haversine distance in miles between two points given their latitudes and longitudes.
    """
    R = 3958.8  # Radius of Earth in miles
 
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
 
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
 
    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance
 
# Apply vectorized function
data['distance_in_miles'] = haversine_vectorized(
    data['lat'].values,
    data['long'].values,
    data['merch_lat'].values,
    data['merch_long'].values
)
# Check the variable
sns.violinplot(x='is_fraud', y='distance_in_miles', data=data)
plt.title('Violin Plot of distance by Fraud Status')
plt.xlabel('Fraud Status (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Distance of transaction from house')
plt.show()
```

```python
# Create an age column and drop the dob column

data['trans_date_trans_time'] = pd.to_datetime(data['trans_date_trans_time'])
data['dob'] = pd.to_datetime(data['dob'])
data['age'] = data['trans_date_trans_time'].dt.year - data['dob'].dt.year

data['age'] -= ((data['trans_date_trans_time'].dt.month < data['dob'].dt.month) |
              ((data['trans_date_trans_time'].dt.month == data['dob'].dt.month) &
               (data['trans_date_trans_time'].dt.day < data['dob'].dt.day))).astype(int)

data = data.drop(columns=['dob'])
```

```python
# Converting trans_date_trans_time to txn_time_of_day, txn_yr, txn_mnth, txn_day
def time_of_day(x):
    if x >=0 and x < 6: return "Night"
    elif x>= 6 and x <= 12: return "Morning"
    elif x> 12 and x <= 15: return "Afternoon"
    elif x> 15 and x <= 20: return "Evening"
    elif x> 20 and x <= 24: return "Night"
        
data["txn_time_of_day"] = data["trans_date_trans_time"].dt.hour.apply(time_of_day)
data["txn_yr"] = data["trans_date_trans_time"].dt.year
data["txn_mnth"] = data["trans_date_trans_time"].dt.month
data["txn_day"] = data["trans_date_trans_time"].dt.day_name()

data = data.drop(columns=['trans_date_trans_time'])
```
